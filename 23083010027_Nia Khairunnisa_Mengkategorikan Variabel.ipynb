# Membuat Visualisasi data untuk Mengkategorikan Distribusi dari Variabel yang Mempengaruhi

```python
# Membuat visualisasi distribusi dari GPA
plt.figure(figsize=(8, 6))
sns.histplot(df['GPA'], kde=True, bins=20, color="skyblue")
plt.title('Distribution of GPA')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

Penjelasan Grafik 
Berdasarkan grafik yang ditampilkan diatas menunjukkan bahwa GPA memiliki bentuk yang simetris menyerupai lonceng, yang merupakan ciri dari distribusi normal.Pada sumbu horizontal, kita melihat nilai GPA yang berkisar antara 0 hingga 4, sementara sumbu vertikal menunjukkan frekuensi atau jumlah siswa yang menerima masing-masing nilai GPA. Puncak tertinggi dari grafik terletak di sekitar nilai GPA 2.5 hingga 3.0, yang menandakan bahwa sebagian besar siswa mendapatkan nilai di kisaran tersebut. Frekuensi kemudian menurun secara bertahap ke arah kedua ujung, yaitu untuk GPA yang lebih rendah dan lebih tinggi.

Fenomena ini mencerminkan pola umum dalam data akademis, di mana banyak siswa cenderung mendapatkan nilai di sekitar rata-rata, dengan jumlah yang semakin sedikit di kedua ekstrem. Dengan demikian, grafik ini menunjukkan bahwa distribusi nilai GPA dalam kelompok ini mendekati distribusi normal, yang sering kali terjadi dalam konteks pendidikan karena variabilitas kemampuan belajar di antara siswa.

```python
# Membuat visualisasi distribusi dari variabel Support Orang tua
plt.figure(figsize=(8, 6))
sns.kdeplot(df['ParentalSupport'], color="yellow", label="Parental Support")
plt.title('Distribution of Parental Support')
plt.xlabel('Parental Support Level')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Bentuk grafik ini menunjukkan bahwa tidak ada satu tingkat dukungan yang dominan. Sebaliknya, terdapat beberapa tingkat yang terlihat cukup sering muncul, menciptakan pola multimodal. Hal ini bisa mencerminkan variasi dalam pengalaman dukungan orang tua di antara individu, yang mungkin disebabkan oleh faktor-faktor seperti latar belakang sosial ekonomi, budaya, atau lokasi geografis.

Selain itu, grafik ini tidak simetris dan tidak mengikuti pola distribusi normal yang biasanya memiliki satu puncak. Keberadaan beberapa puncak menunjukkan bahwa responden memiliki pengalaman yang berbeda-beda terkait dukungan orang tua, yang membuat distribusi ini lebih kompleks. Dengan demikian, kita bisa menyimpulkan bahwa distribusi tingkat dukungan orang tua ini adalah multimodal, mencerminkan keragaman dalam pengalaman individu terhadap dukungan yang diterima dari orang tua mereka.

# Pemusatan Data

```python
# Pemusatan data pada variabel faktor yang berpengaruh
kolom = ['ParentalEducation','StudyTimeWeekly','Absences','Tutoring','ParentalSupport','Extracurricular','Sports','Music','Volunteering','GPA','GradeClass']
hasil = []
for perhitungan in kolom:
    mean = df[perhitungan].mean()
    median = df[perhitungan].median()
    modus = df[perhitungan].mode()[0]
    
    hasil.append({
        'Faktor Pengaruh': perhitungan,
        'Mean': mean,
        'Median': median,
        'Modus': modus
    })
    
hasil_df = pd.DataFrame(hasil)
print(hasil_df)
```

Perhitungan di atas merupakan perhitungan dari faktor yang mempengaruhi GPA siswa yaitu pencarian rata-rata, nilai median dan modusnya dari variabel faktor pengaruh

# Penyebaran Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Definisikan ulang kolom numerik
kolom_numerik = ['ParentalEducation','StudyTimeWeekly','Absences','Tutoring','ParentalSupport','Extracurricular','Sports','Music','Volunteering','GPA','GradeClass']

# Inisialisasi list untuk menyimpan hasil penyebaran
hasil_sebar = []

# Menghitung penyebaran untuk setiap kolom
for perhitungan in kolom_numerik:
    rentang = df[perhitungan].max() - df[perhitungan].min()
    varians = df[perhitungan].var()
    deviasi_standar = df[perhitungan].std()
    q1 = df[perhitungan].quantile(0.25)  # Kuartil 1
    q2 = df[perhitungan].median()        # Kuartil 2 (median)
    q3 = df[perhitungan].quantile(0.75)  # Kuartil 3
    
    hasil_sebar.append({
        'Faktor Pengaruh': perhitungan,
        'Range': rentang,
        'Varians': varians,
        'Standar Deviasi': deviasi_standar,
        'Q1': q1,
        'Q2': q2,
        'Q3': q3
    })

# Konversi hasil ke DataFrame untuk tampilan yang lebih baik
hasil_sebar_df = pd.DataFrame(hasil_sebar)
print(hasil_sebar_df)

# Plot penyebaran data menggunakan box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[kolom_numerik])
plt.title('Penyebaran Data untuk Variabel Numerik')
plt.xlabel('Variabel')
plt.ylabel('Nilai')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Definisikan ulang kolom numerik
kolom_numerik = ['Age', 'ParentalEducation', 'StudyTimeWeekly', 'Absences', 'GPA', 'GradeClass']

# Plot untuk setiap kolom
for perhitungan in kolom_numerik:
    # Buat plot distribusi menggunakan KDE plot dengan isian di bawah garis
    plt.figure(figsize=(6, 4))
    sns.kdeplot(df[perhitungan], color='pink', fill=True, alpha=0.5)  # KDE plot dengan arsiran (fill=True)
    
    # Hitung mean, median, modus, dan skewness
    mean = df[perhitungan].mean()
    median = df[perhitungan].median()
    modus = df[perhitungan].mode()[0]
    skewness = skew(df[perhitungan])
    
    # Tambahkan garis mean, median, dan modus pada plot
    plt.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='green', linestyle='-', label=f'Median: {median:.2f}')
    plt.axvline(modus, color='blue', linestyle='-', label=f'Modus: {modus:.2f}')
    
    # Tambahkan label dan judul
    plt.title(f'Skewness Distribution of {perhitungan} (Skewness: {skewness:.2f})')
    plt.xlabel(perhitungan)
    plt.ylabel('Frekuensi')
    plt.legend()
    
    # Tampilkan hasil analisis skewness
    if skewness > 0:
        print(f"{perhitungan}: Distribusi condong ke kanan (right-skewed).")
    elif skewness < 0:
        print(f"{perhitungan}: Distribusi condong ke kiri (left-skewed).")
    else:
        print(f"{perhitungan}: Distribusi simetris.")
    
    # Tampilkan plot
    plt.show()
```
